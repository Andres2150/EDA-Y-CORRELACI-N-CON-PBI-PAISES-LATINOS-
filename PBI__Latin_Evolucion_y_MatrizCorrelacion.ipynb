```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

# --- DATOS (Utilizaremos el mismo DataFrame df_absolute del análisis anterior) ---

countries = ['BRASIL', 'MEXICO', 'ARGENTINA', 'COLOMBIA', 'CHILE', 'PERU', 'ECUADOR', 'REP DOM.', 'GUATEMALA', 'COSTA RICA', 'PANAMA', 'URUGUAY', 'BOLIVIA', 'PARAGUAY', 'HONDURAS', 'EL SALVADOR', 'NICARAGUA']

df_absolute = pd.DataFrame(data, index=countries).T
df_absolute.index = pd.to_numeric(df_absolute.index)
df_absolute = df_absolute.apply(pd.to_numeric, errors='coerce')

# Definir el grupo 1
grupo_1_paises = ['BRASIL', 'MEXICO', 'ARGENTINA', 'COLOMBIA', 'PERU', 'CHILE']
df_g1 = df_absolute[grupo_1_paises]
final_year = 2024

# Función para formatear el eje Y a Billones (T)
def millions_formatter(x, pos):
    if x >= 1000:
        return f'{x/1000:.1f}T'
    return f'{x:.0f}'

# --- Generación del Gráfico 1 ---
plt.figure(figsize=(12, 7))
sns.set_palette("tab10")

for country in grupo_1_paises:
    plt.plot(df_g1.index, df_g1[country], linewidth=2, label=country)

    final_value = df_g1[country].iloc[-1]

    # Etiquetas flotantes
    label_text = f"{country} ({final_value / 1000:.2f}T)"
    plt.text(
        final_year + 0.1,
        final_value,
        label_text,
        fontsize=10,
        fontweight='bold',
        color=plt.gca().lines[-1].get_color()
    )

plt.title('Gráfico 1: PBI Absoluto - Las 6 Economías Más Grandes (2000-2024)', fontsize=16)
plt.xlabel('Año')
plt.ylabel('PBI (Miles de Millones de USD)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xlim(2000, 2025.5)
plt.gca().yaxis.set_major_formatter(FuncFormatter(millions_formatter))
plt.show()

print("")
```

```python
# --- GRÁFICO 2: ANDINOS Y EL RESTO ---
# Países a incluir: Colombia, Perú, Chile y todos los demás (excluyendo a Brasil y México)
grupo_2_paises = ['COLOMBIA', 'PERU', 'CHILE', 'ECUADOR', 'REP DOM.', 'GUATEMALA', 'COSTA RICA', 'PANAMA', 'URUGUAY', 'BOLIVIA', 'PARAGUAY', 'HONDURAS', 'EL SALVADOR', 'NICARAGUA']
df_g2 = df_absolute[grupo_2_paises]

# --- Generación del Gráfico 2 ---
plt.figure(figsize=(12, 7))
sns.set_palette("tab20") # Usar una paleta más amplia

for country in grupo_2_paises:
    # Usar un grosor de línea ligeramente menor para el grupo más grande de países
    linewidth = 2.5 if country in ['COLOMBIA', 'PERU', 'CHILE'] else 1.5
    plt.plot(df_g2.index, df_g2[country], linewidth=linewidth, label=country)

    final_value = df_g2[country].iloc[-1]

    # Etiquetas flotantes
    # Mostrar el valor en miles de millones (x.x MM) para los países más pequeños para mayor claridad
    if final_value >= 100:
        label_text = f"{country} ({final_value:.0f} MM)" # Cientos de miles de millones
    else:
        label_text = country # Deja el nombre si es muy pequeño

    plt.text(
        final_year + 0.1,
        final_value,
        label_text,
        fontsize=8,
        fontweight='bold',
        color=plt.gca().lines[-1].get_color()
    )

plt.title('Gráfico 2: PBI Absoluto - Andinos y el Resto de Latinoamérica (2000-2024)', fontsize=16)
plt.xlabel('Año')
plt.ylabel('PBI (Miles de Millones de USD)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.xlim(2000, 2025.5)
plt.show()

print("")
```

```python


#***********************************************************
#Matriz de Correlación del PBI entre Países (2000-2024)
#***********************************************************

df = pd.DataFrame(data, index=countries).T #Convierte tus datos en una tabla (DataFrame)
df = df.apply(pd.to_numeric, errors='coerce')#Convierte todo el DataFrame a números reales (int o float).


# --- 2. CÁLCULO DE LA CORRELACIÓN Y CREACIÓN DE LA MÁSCARA ---

# Calcular la Matriz de Correlación de Pearson
# Nota: La correlación se calcula sobre las series de tiempo (eje 0), por lo que debemos transponer los datos.
corr_matrix = df.corr(method='pearson')

# Crear la Máscara Triangular Superior
# np.triu(np.ones_like(corr_matrix)) crea una matriz de unos con la forma de la matriz de correlación
# y establece en cero los elementos de la matriz triangular inferior.
# k=1 excluye la diagonal principal (donde todos los valores son 1)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool), k=1)


# --- 3. GENERACIÓN DEL GRÁFICO TRIANGULAR ---

plt.figure(figsize=(12, 10))

# Crear el mapa de calor (heatmap) aplicando la máscara
sns.heatmap(
    corr_matrix,
    mask=mask,               # Aplicar la máscara para ocultar la mitad superior
    annot=True,              # Mostrar los valores de correlación
    fmt=".2f",               # Formato a dos decimales
    cmap='coolwarm',         # Usar un mapa de color divergente (como en tu imagen)
    linewidths=.5,           # Líneas entre las celdas
    cbar_kws={"shrink": .8, "label": "Coeficiente de Correlación de Pearson (r)"} # Ajustar la barra de color
)

plt.title('Matriz de Correlación del PBI entre Países (2000-2024)', fontsize=16)

##ANALISIS DE LA CORRELACION COMPARADA
#Sub-bloque muy fuerte: economías medianas–pequeñas
#Países como:#Perú,Ecuador,Panamá,Costa Rica,Rep. Dominicana,Guatemala,Nicaragua,Honduras,El Salvador,
#presentan correlaciones cercanas a 0.97–0.99 entre sí.
#Estas economías comparten un patrón de crecimiento casi idéntico.

#Lectura estructural, Economías abiertas,
#Dependientes de:Exportaciones primarias, Remesas, Turismo
#Reaccionan igual ante shocks externos (2009, 2020)

#Brasil: el país más “independiente” del sistema
#CONCLUSION:  Brasil sigue su propio ciclo económico.
#Razones: Mercado interno enorme, Política macro más autónoma, Menor dependencia relativa de remesas y turismo

plt.yticks(rotation=0)      # Mantener las etiquetas del eje Y horizontales
plt.show()

print("")
```
